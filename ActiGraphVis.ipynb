```python
import pyActigraphy
import plotly.graph_objects as go
raw = pyActigraphy.io.read_raw_agd('Left Leg (2021-07-13)30sec.agd')
```

```python
import pandas as pd

df = pd.DataFrame(list(zip(raw.data.index.astype(str), raw.data, raw1.data, raw2.data)),
               columns =['x', 'y1', 'y2', 'y3'])
```

```python
import matplotlib.pyplot as plt
ax = plt.gca()
df.plot(kind='line',x='x',y='y1',color='red', ax=ax, linestyle='dashed',  linewidth=1)
df.plot(kind='line',x='x',y='y2',color='blue', ax=ax, linestyle='dashed',  linewidth=1)
df.plot(kind='line',x='x',y='y3',color='green', ax=ax, linestyle='dashed',  linewidth=1)
plt.show()
```
